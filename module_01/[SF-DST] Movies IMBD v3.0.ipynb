```python
# Не забываем вопрос 6 - там создали колонку profit

# Выберем где фильмы с прибылью
data_profit = data[data.profit > 0]

# Поместим все в функцию, так как это нам понадобится далее
# Бежим по всем индексам, в текстовой колонке рассплитовываем по "|" и собираем в список
def text_col_split(text_col, df):
    text_all = []
    for num in df.index:
        row = df.loc[num][text_col]
        for i in row.split("|"):
            text_all.append(i)
    return text_all

# создаем список, где все упоминания жанров с повторами со всех прибыльных
genres_data_profit = text_col_split('genres', data_profit)
#display(len(genres_data_profit))

# считаем по наличию
genres_data_profit_count = Counter()
for gen in genres_data_profit:
    genres_data_profit_count[gen] += 1
    
# .most_common(3) показать первые три самых больших
display(genres_data_profit_count.most_common(3))



answer_ls.append(1)
```

# 31. Названия фильмов какой студии в среднем самые длинные по количеству символов?
Варианты ответа:
1. Universal Pictures (Universal)
2. Warner Bros
3. Jim Henson Company, The
4. Paramount Pictures
5. Four By Two Productions

```python
# словарь с длинами имен фильмов
film_name_len_dict = {}
# словарь с количеством фильмов студии
film_name_count_dict = {}

for num in range(len(data)):
    # строка с производственными компаниями
    row = data.iloc[num]['production_companies']
    for i in row.split("|"):
        #original_title_len - длина названия фильма
        original_title_len = len(data.loc[num]['original_title'])
        film_name_len_dict.setdefault(i, 0)
        # сюда суммируем все длины названий фильмов по студии
        film_name_len_dict[i] = film_name_len_dict[i]+original_title_len
        # сюда суммируем все фильмы по студии
        film_name_count_dict.setdefault(i, 0)
        film_name_count_dict[i] = film_name_count_dict[i]+1        
#print(film_name_len_dict)
#print(film_name_count_dict)

# заводим словарь, где будем делить длину названий на количество фильмов
film_name_len_mean_dict = {}
for k, v in film_name_len_dict.items():
    if k  in film_name_count_dict:
        film_name_len_mean_dict.setdefault(k, 0)
        film_name_len_mean_dict[k] = film_name_len_dict[k]/film_name_count_dict[k]

max_mean = max(list(film_name_len_mean_dict.values()))

for k, v in film_name_len_mean_dict.items():
    if v == max_mean:
        print(k, v)


answer_ls.append(5)
```
